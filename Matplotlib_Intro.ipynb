# Introduction to Matplotlib

```python
from matplotlib import style
import matplotlib.pyplot as plt
```

## Very Basic Plot

```python
plt.plot([5,5,20,20],[5,40,5,40])
plt.title("Initial Letter")
plt.show()
```

```python
# visualizing two lines
x = [1,2,3]
y = [4,5,6]
x1 = [4,5,6]
y1 = [1,2,3]
plt.plot(x,y)
plt.plot(x1,y1)
plt.title('Just Test')
plt.xlabel('test x values')
plt.ylabel('test y values')
plt.show()
```

## Plotting Bar chart 

```python
x = [1,2,3]
y = [4,5,6]

x1 = [3,2,1]
y1 = [4,6,5]

x2 = [2,1,4]
y2 = [5,3,6]

style.use('classic')
plt.bar(x,y)
plt.bar(x1,y1)
plt.bar(x2,y2)
plt.title('test graph 1')
plt.xlabel('test x name')
plt.ylabel('test y name')
plt.show()
```

## Histograms

```python
style.use('bmh')
numbers = [1,2,4,6,3,62,6,72,53,46,2,1,45,6,3,6,36,7,56,76,35,67,31,54,3,22,66]
jumps = [0,2,3,5,6,7,8,9]
plt.hist(numbers,jumps,histtype = 'bar')
plt.hist(numbers,jumps,histtype = 'step')
plt.title('test hist 1')
plt.xlabel('test xlabel')
plt.ylabel('test ylabel')
plt.show()
```

## Pie Chart

```python
style.use('classic')

foods = ['pizza', 'Ice cream', 'burgers']
sales = [28,10,25]
color = ['red','blue','yellow']
plt.pie(sales,labels = foods,colors = color)
plt.title('example food')
plt.show()
```

## Scatter plot

```python
import numpy as np

x = np.linspace(1,2,20).reshape(5,4)
y = np.random.random(20)
print(y)
plt.scatter(x,y)
plt.show()
```

```python
a = np.array([[1,2,3,4],[5,6,7,8]])
plt.plot(a[0],a[1],'go')
plt.plot(a[0],a[1],'--',color='red')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

## Subplots

```python
t = np.linspace(0,5,20)
#plt.plot(t,t**2,color = 'coral', linestyle = '--')
#plt.plot(t,4*t,color = 'indigo',linewidth = '2')

plt.subplot(1,4,1)
plt.plot(t,t,'c--')

plt.subplot(1,4,2)
plt.plot(t,t**2,'b--')
plt.subplot(1,4,3)
plt.plot(t,t**3,'mo')
plt.subplot(1,4,4)
plt.plot(t,t**4)
```

## Bar graphs

```python
grades = {'john':70,
         'susan':80,
         'ben':92,
         'david':98,
         'alex':76}

plt.bar(range(len(grades)),list(grades.values()),color = 'darkturquoise')
plt.xticks(range(len(grades)),list(grades.keys()))
plt.xlabel('students')
plt.ylabel('grades')
plt.show()
```
