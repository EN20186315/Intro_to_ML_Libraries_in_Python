## Creating dataframe form dictionary

```python
employee = {'number':[1,2,3,4],
           'name':['AP','NK','VP','NC'],
           'hourly salary':[15,35,74,40]}

# print(employee)

# Creating table using dataframe()
table = pd.DataFrame(employee)
print(table)

print("\n",table.head(2))
print("\n",table.tail(2))
print("\n",table.tail(1))
```

## Using merge() function

This function merges same items

```python
# Assuming dataset
food = {'number': [1,2,3,4,5],
       'name': ['apple','banana','chocolate','chips','almonds'],
       'price': [2,4,5,72,34]}

food1 = {'number': [1,2,3,4,5],
       'name': ['apple','banana','chocolate','chips','almonds'],
       'price': [2,4,5,72,34]}

# Creating tables 
table = pd.DataFrame(food)
table1 = pd.DataFrame(food1)

# using merge() function 
fusion = pd.merge(table,table1)
print(fusion)
fusion1 = pd.merge(table,table1, on = 'number')
print("\n",fusion1)

fusion2 = pd.merge(table,table1, on = 'name')
print("\n",fusion2)
```

## Using Join function to join two dataframes or datasets.

```python
food = {'number': [1,2,3,4,5],
       'name': ['apple','banana','chocolate','chips','almonds'],
       'price': [2,4,5,72,34]}

food1 = {'color' : ['red','yellow','orange','blue','indigo'],
        'weight' : [123,536,234,65,1234],
        'qt' : [2,4,5,1,65]}

table = pd.DataFrame(food)
table1 = pd.DataFrame(food1)

# Using join() function
fusion = table.join(table1)
print(fusion)
```

##  Creating Series

```python
# Creating Series from dict.
d = {'Banglore' : 15000,
    'Pune' : 12342,
    'Delhi' : 13456,
    'Kolhapur' : 10234}

cities = pd.Series(d)
print(cities)

print("\n",cities[['Banglore','Kolhapur']])
print("\n",cities[cities<13000])
```
